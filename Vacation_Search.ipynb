```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Store the CSV you saved created in part one into a DataFrame.
city_data_df = pd.read_csv("weather_data/WeatherPy_challenge.csv", index_col=0)
city_data_df.head()
```

```python
# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
rain_ques = input("Do you want it to be raining? (yes/no) ")
snow_ques = input("Do you want it to be snowing? (yes/no) ")
```

```python
# Filter the dataset to find the cities that fit the criteria.
temp_cities_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & \
                                       (city_data_df["Max Temp"] >= min_temp)]
temp_cities_df
```

```python
# Use if statemeants to find where it is raining or snowing.
if (rain_ques == "no") & (snow_ques == "no"):
    preferred_cities_df = temp_cities_df.loc[(temp_cities_df["Rain Inches (last 3 hrs)"] == 0) & \
                                       (temp_cities_df["Snow Inches (last 3 hrs)"] == 0)]
    
elif (rain_ques == "yes") & (snow_ques == "no"):
    preferred_cities_df = temp_cities_df.loc[(temp_cities_df["Rain Inches (last 3 hrs)"] != 0) & \
                                       (temp_cities_df["Snow Inches (last 3 hrs)"] == 0)]
    
elif (rain_ques == "no") & (snow_ques == "yes"):
    preferred_cities_df = temp_cities_df.loc[(temp_cities_df["Rain Inches (last 3 hrs)"] == 0) & \
                                       (temp_cities_df["Snow Inches (last 3 hrs)"] != 0)]

else: 
    preferred_cities_df = temp_cities_df.loc[(temp_cities_df["Rain Inches (last 3 hrs)"] != 0) & \
                                       (temp_cities_df["Snow Inches (last 3 hrs)"] != 0)]
```

```python
# Drop any null rowns
preferred_cities_df = preferred_cities_df.dropna()

preferred_cities_df.head(10)
```

```python
preferred_cities_df
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Set parameters to search for a hotel.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key}
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try: hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
#Create markers
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.
locations = hotel_df[["Lat", "Lng"]]
# max_temp = hotel_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
# heat_layer = gmaps.heatmap_layer(locations, weights=max_temp,dissipating=False,
#              max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
# fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```

```python
# Create the output file (CSV).
output_data_file = "weather_data/WeatherPy_vacation.csv"
# Export the City_Data into a CSV.
hotel_df.to_csv(output_data_file)
```
